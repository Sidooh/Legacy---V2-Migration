```python
import pandas as pd
```

```python
users = pd.read_csv("old/sidooh_users.csv")
accounts = pd.read_csv("old/sidooh_accounts.csv", dtype={'phone': str, 'pin': str, 'user_id': str, 'referrer_id': str},)
invites = pd.read_csv("old/sidooh_referrals.csv", dtype={'referee_phone': str, 'account_id': str, 'referee_id': str},)
```

```python
users_new = users.copy()
accounts_new = accounts.copy()
invites_new = invites.copy()
```

```python
users_new['username'] = users['username'].apply(str).apply(lambda value: value[-12:-1])
users_new['id_number'] = users['id_number'].apply(str).apply(lambda value: value[-12:-1])
users_new['password'] = users['password'].apply(lambda value: '')
del users_new['remember_token']

users_new = users_new.drop(users_new.index[[37, 39, 99, 293]])
```

```python
users_new
```

```python
accounts_new['inviter_id'] = accounts['referrer_id']
del accounts_new['referrer_id']

accounts_new.at[61, 'user_id'] = 36
accounts_new.at[62, 'user_id'] = 39
accounts_new.at[222, 'user_id'] = 16
accounts_new.at[599, 'user_id'] = 48
```

```python
accounts_new
```

```python
invites_new['inviter_id'] = invites['account_id']
invites_new['account_id'] = invites['referee_id']
invites_new['phone'] = invites['referee_phone']
del invites_new['referee_id']
del invites_new['referee_phone']
```

```python
users_new.to_csv('new/accounts/sidooh_users.csv', index = False)
accounts_new.to_csv('new/accounts/sidooh_accounts.csv', index = False)
invites_new.to_csv('new/accounts/sidooh_invites.csv', index = False)
```
