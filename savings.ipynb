```python
import pandas as pd
```

```python
sub_accounts = pd.read_csv("old/sidooh_sub_accounts.csv")
sub_investments = pd.read_csv("old/sidooh_sub_investments.csv")
collective_investments = pd.read_csv("old/sidooh_collective_investments.csv")
```

```python
sub_accounts_new = sub_accounts.copy()
sub_investments_new = sub_investments.copy()
collective_investments_new = collective_investments.copy()
```

```python
collective_investments_new['invested_at'] = collective_investments_new['investment_date']
del collective_investments_new['investment_date']
```

```python
sub_investments_new = sub_investments.copy()

dateFrom = pd.to_datetime('now') + pd.DateOffset(months=-1)
sub_investments_new = sub_investments_new[pd.to_datetime(sub_investments_new['created_at']) > dateFrom]
sub_investments_new = sub_investments_new[sub_investments_new['interest'] > 0]
sub_investments_new = sub_investments_new[sub_investments_new['account_id'] % 3 > 0]
sub_investments_new.reset_index(inplace=True, drop=True)

sub_investments_new['personal_account_id'] = sub_investments_new['account_id']
sub_investments_new['personal_collective_investment_id'] = sub_investments_new['collective_investment_id']

del sub_investments_new['account_id']
del sub_investments_new['collective_investment_id']

# sub_investments_new.iloc[6370:6400]
```

```python
def typeTransform(val):    
    if (val == "SAVINGS"):
        val = "LOCKED"
        
    return val

sub_accounts_new['type'] = sub_accounts['type'].apply(typeTransform)


interest_acc_mask = sub_accounts_new['type'] == 'INTEREST'
interest_sub_accs = sub_accounts_new[interest_acc_mask]
main_sub_accs = sub_accounts_new[~interest_acc_mask]

main_sub_accs['description'] = ''
main_sub_accs['status'] = 'ACTIVE'
main_sub_accs['balance'] = sub_accounts['in'] - sub_accounts['out']

main_sub_accs
```

```python
interest_sub_accs
```

```python
interest_sub_accs_current = interest_sub_accs.copy()
interest_sub_accs_locked = interest_sub_accs.copy()

interest_sub_accs_current['in'] = interest_sub_accs['in'].apply(lambda val: val *.2)
interest_sub_accs_locked['in'] = interest_sub_accs['in'] - interest_sub_accs_current['in']

interest_sub_accs_current['out'] = interest_sub_accs['out'].apply(lambda val: val *.2)
interest_sub_accs_locked['out'] = interest_sub_accs['out'] - interest_sub_accs_current['out']

interest_sub_accs_current['balance'] = interest_sub_accs_current['in'] - interest_sub_accs_current['out']
interest_sub_accs_locked['balance'] = interest_sub_accs_locked['in'] - interest_sub_accs_locked['out']

interest_sub_accs_current['type'] = "CURRENT"
interest_sub_accs_locked['type'] = "LOCKED"


interest_sub_accs_joined = pd.concat([interest_sub_accs_current, interest_sub_accs_locked]).sort_values('account_id')

del interest_sub_accs_joined['id']
del interest_sub_accs_joined['created_at']
del interest_sub_accs_joined['updated_at']

interest_sub_accs_joined
```

```python
sub_accounts_rejoined = main_sub_accs.join(interest_sub_accs_joined.set_index(['account_id', 'type']), on=['account_id', 'type'], rsuffix="_interest")
sub_accounts_rejoined
```

```python
sub_accounts_final = sub_accounts_rejoined.copy()

sub_accounts_final['interest'] = sub_accounts_final['balance_interest']
sub_accounts_final[['deleted_at', 'frequency', 'duration']] = ''

sub_accounts_final.drop(columns=['in', 'out', 'balance_interest', 'in_interest', 'out_interest'], inplace=True)


sub_accounts_final
```

```python
personal_account_trans = sub_accounts_rejoined.copy()
personal_account_trans[['created_at', 'updated_at']] = pd.to_datetime('now')

personal_account_trans['personal_account_id'] = personal_account_trans['id']
personal_account_trans.drop(columns=['id', 'in', 'out', 'balance_interest', 'in_interest', 'account_id'], inplace=True)

personal_account_trans = personal_account_trans[personal_account_trans['balance'] > 0]

personal_account_trans_in = personal_account_trans.copy()

personal_account_trans_in['type'] = 'CREDIT'
personal_account_trans_in['status'] = 'COMPLETED'
personal_account_trans_in['amount'] = personal_account_trans_in['balance']
personal_account_trans_in['description'] = "Savings B/F"

del personal_account_trans_in['balance']
del personal_account_trans_in['out_interest']

personal_account_trans_interest = personal_account_trans.copy()

personal_account_trans_interest = personal_account_trans_interest[personal_account_trans_interest['out_interest'] > 0]

personal_account_trans_interest['type'] = 'CREDIT'
personal_account_trans_interest['status'] = 'COMPLETED'
personal_account_trans_interest['amount'] = personal_account_trans_interest['out_interest']
personal_account_trans_interest['description'] = "Interest B/F"

del personal_account_trans_interest['balance']
del personal_account_trans_interest['out_interest']


personal_account_trans
```

```python
personal_account_transactions = pd.concat([personal_account_trans_in, personal_account_trans_interest]).sort_values('personal_account_id')
personal_account_transactions
```
